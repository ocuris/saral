### рдСрдкрд░реЗрдЯрд░реНрд╕ рдХреНрдпрд╛ рд╣реИрдВ?

* рдСрдкрд░реЗрдЯрд░реНрд╕ = рдЧрдгрд┐рдд рдХрд░рдиреЗ рдХреЗ рдЙрдкрдХрд░рдгред
* рдпреЗ рд╣рдореЗрдВ рдЬреЛрдбрд╝, рдШрдЯрд╛рд╡, рдЧреБрдгрд╛, рднрд╛рдЧ рдФрд░ рдмрд╛рдХреА рдЧрдгрдирд╛рдПрдБ рдХрд░рдиреЗ рджреЗрддреЗ рд╣реИрдВред

рдкрд╛рдпрдерди рдореЗрдВ рд╕рдмрд╕реЗ рдЬрд╝реНрдпрд╛рджрд╛ рдЗрд╕реНрддреЗрдорд╛рд▓ рд╣реЛрдиреЗ рд╡рд╛рд▓реЗ **Arithmetic Operators** рдпреЗ рд╣реИрдВ:

| рдСрдкрд░реЗрдЯрд░ | рдХрд╛рдо |
|---------|-----|
| `+` | рдЬреЛрдбрд╝ (Addition) |
| `-` | рдШрдЯрд╛рд╡ (Subtraction) |
| `*` | рдЧреБрдгрд╛ (Multiplication) |
| `/` | рднрд╛рдЧ (Division - рдкрд░рд┐рдгрд╛рдо рджрд╢рдорд▓рд╡ рдореЗрдВ) |
| `//` | рднрд╛рдЧ (Division - рдХреЗрд╡рд▓ рдкреВрд░реНрдгрд╛рдВрдХ) |
| `%` | рд╢реЗрд╖рдлрд▓ (Remainder) |
| `**` | рдШрд╛рддрд╛рдВрдХ (Power) |

````python
# рдЙрджрд╛рд╣рд░рдг: рдореЛрд╣рди рдХреА рдмреЛрд░рд┐рдпреЛрдВ рдХрд╛ рд╣рд┐рд╕рд╛рдм
monday = 10
tuesday = 12
wednesday = 8

# рдЬреЛрдбрд╝
print("рд╕реЛрдорд╡рд╛рд░ + рдордВрдЧрд▓рд╡рд╛рд░ =", monday + tuesday)

# рдШрдЯрд╛рд╡
print("рдордВрдЧрд▓рд╡рд╛рд░ - рдмреБрдзрд╡рд╛рд░ =", tuesday - wednesday)

# рдЧреБрдгрд╛
print("рд╕реЛрдорд╡рд╛рд░ ├Ч 2 =", monday * 2)

# рднрд╛рдЧ (decimal рдореЗрдВ)
print("рдордВрдЧрд▓рд╡рд╛рд░ ├╖ 3 =", tuesday / 3)

# рднрд╛рдЧ (integer рдореЗрдВ)
print("рдордВрдЧрд▓рд╡рд╛рд░ ├╖ 3 (рдкреВрд░реНрдгрд╛рдВрдХ) =", tuesday // 3)

# рд╢реЗрд╖рдлрд▓
print("рдордВрдЧрд▓рд╡рд╛рд░ ├╖ 3 рдХрд╛ рд╢реЗрд╖рдлрд▓ =", tuesday % 3)

# рдШрд╛рддрд╛рдВрдХ
print("2 рдХреА рдШрд╛рдд 3 =", 2 ** 3)
````
